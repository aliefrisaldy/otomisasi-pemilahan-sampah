```python
from ultralytics import YOLO

# 1. Load model dasar (pilih yang Nano 'n' agar ringan dan cepat di RTX 2050)
model = YOLO('yolov8s.pt')

# 2. Definisikan lokasi file data.yaml
# Ganti path di bawah ini dengan path LENGKAP ke file data.yaml yang baru kamu edit
path_to_yaml = r"D:\grinding\project\Otomisasi Pemilahan Sampah Berbasil Visi Komputer\dataset\YOLO-Waste-Detection-1\YOLO-Waste-Detection-1\data.yaml"

# Mulai training
results = model.train(
    data=path_to_yaml,   # Pastikan path ke file yaml dataset benar
    epochs=50,         # Rencana baru Anda
    imgsz=640,          # Resolusi standar
    batch=-1,            # DITURUNKAN agar muat di RTX 2050 4GB
    workers=2,          # Jumlah worker data loading
    patience=20,        # Early Stopping: Stop jika tidak ada perbaikan selama 20 epoch
    name='yolov8s_trash_sorting', # Nama folder hasil agar rapi
    augment=True,       # Aktifkan augmentasi default
    lr0=0.01,           # Initial learning rate standar
    optimizer='auto' #Alif 
)
```
